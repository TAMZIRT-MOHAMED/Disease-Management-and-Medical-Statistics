```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data
data = {'Gender': ['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'],
        'Diagnosis': ['Flu', 'Cold', 'Flu', 'COVID', 'Cold', 'Flu', 'COVID', 'COVID', 'Flu', 'Cold']}
df = pd.DataFrame(data)

# Count occurrences
pivot_df = df.pivot_table(index="Diagnosis", columns="Gender", aggfunc="size", fill_value=0)

# Normalize for proportion (percentage)
pivot_df_percentage = pivot_df.div(pivot_df.sum(axis=1), axis=0) * 100  # Convert to percentage

# Plot
ax = pivot_df_percentage.plot(kind="bar", stacked=True, figsize=(8,5), colormap="coolwarm")

# Formatting
plt.title("Proportion of Diagnoses by Gender")
plt.xlabel("Diagnosis")
plt.ylabel("Percentage")
plt.legend(title="Gender")
plt.xticks(rotation=0)  # Keep labels horizontal
plt.ylim(0, 100)  # Y-axis from 0 to 100%

# Show plot
plt.show()
```

```python
import tkinter as tk

def create_gradient(canvas, width, height, start_color, end_color):
    """
    Creates a vertical gradient on the given canvas.
    """
    r1, g1, b1 = canvas.winfo_rgb(start_color)  # Start color
    r2, g2, b2 = canvas.winfo_rgb(end_color)  # End color
    canvas.pack_propagate(False)
    r_ratio = (r2 - r1) / height
    g_ratio = (g2 - g1) / height
    b_ratio = (b2 - b1) / height

    for i in range(height):
        r = int(r1 + (r_ratio * i))
        g = int(g1 + (g_ratio * i))
        b = int(b1 + (b_ratio * i))
        color = f"#{r//256:02x}{g//256:02x}{b//256:02x}"
        canvas.create_line(0, i, width, i, fill=color, width=1)

# Create Tkinter window
root = tk.Tk()
root.geometry("400x300")

# Create a Canvas for the gradient
canvas = tk.Canvas(root, width=400, height=300)
canvas.pack(fill="both", expand=True)
# Apply gradient effect
create_gradient(canvas, 400, 300, "#ff0000", "#0000ff")  # Red to Blue

root.mainloop()
```

```python
# import tkinter as tk
# root = tk.Tk()
# root.title("Canvas Example")

# # Create a canvas with width=400 and height=300
# canvas = tk.Canvas(
#     root,
#     width=400, 
#     height=300, 
#     bg="Green" , 
#     relief="flat"  # relief="flat" is default , relief="raised" is for 3D effect and relief="sunken" is for pressed effect , relief="ridge" is for 3D effect and relief="groove" is for 3D effect , relief="solid" is for 3D effect 
#     )
# canvas.pack()
# root.
# root.mainloop()
import tkinter as tk

# Initialize main window
root = tk.Tk()
root.title("Tkinter Animation")

# Create Canvas
canvas = tk.Canvas(root, width=400, height=300, bg="white")
canvas.pack()

# Create a moving object (ball)
ball = canvas.create_oval(10, 10, 50, 50, fill="red")

dx = 5  # Change in x-direction
dy = 3  # Change in y-direction

def move_ball():
    global dx, dy
    canvas.move(ball, dx, dy)
    
    # Get the new position
    x1, y1, x2, y2 = canvas.coords(ball)
    
    # Bounce off walls
    if x1 <= 0 or x2 >= 400:
        dx = -dx
    if y1 <= 0 or y2 >= 300:
        dy = -dy
    
    root.after(30, move_ball)
    root.after(30, move_ball)
    root.after(30, move_ball)  # Re-run after 30ms

# Start animation
move_ball()
root.mainloop()

```

```python
class  Complex:
    def __init__(self, a:float, b:float):
        self.real = a 
        self.imaginary = b 
        self.t()
    def __add__(self, other):
        self.real += other.real
        self.imaginary += other.imaginary
        return self
    def __str__(self):
        return f'{self.real}+ i{self.imaginary}'
    def t(self): pass

a = Complex(1,2)
b = Complex(2,3)

print(a + b )
```

```python
import tkinter as tk
from  pathlib import Path
from  tkinter import Canvas, Button, PhotoImage, StringVar, messagebox
from  fonctionality.PlaceHolder import PlaceholderEntry 

class RegistrationPage(tk.Frame):
    def __init__(self, parent, controller):
        super().__init__(parent, bg="#CBD4F5")
        self.parent = parent
        self.controller = controller
        self.assets_path = Path(Path(str(Path.cwd()) +'/build1/assets/frame1'))
        self.var_full_name = StringVar()
        self.var_User_Name = StringVar()
        self.var_email = StringVar()
        self.var_password = StringVar()
        self.var_confirm_password = StringVar()

        self.create_widgets()

    def relative_to_assets(self, path: str) -> Path:
        return self.assets_path / Path(path)

    def registration_check(self):
        registration_info = {
            "Full Name": self.var_full_name.get(),
            "Username": self.var_User_Name.get(),
            "Email": self.var_email.get(),
            "Password": self.var_password.get(),
            "Confirm Password": self.var_confirm_password.get()
        }

        text = "\n".join([f"{key}: {value}" for key, value in registration_info.items()])
        confirmation = messagebox.askyesnocancel("Confirm Registration", text)
        if confirmation:
            self.controller.show_frame("LoginPage")

    def create_widgets(self):
        canvas = Canvas(
            self,
            bg="#CBD4F5",
            height=600,
            width=1000,
            bd=0,
            highlightthickness=0,
            relief="ridge"
        )
        canvas.place(x=0, y=0)

        self.image_image_1 = PhotoImage(
            file=self.relative_to_assets("image_1.png"))
        self.image_1 = canvas.create_image(
            34.0,
            34.0,
            image=self.image_image_1
        )
        self.image_image_2 = PhotoImage(
        file=self.relative_to_assets("image_2.png"))
        self.image_2 = canvas.create_image(
            297.0,
            237.0,
            image=self.image_image_2
        )

        self.image_image_3 = PhotoImage(
            file=self.relative_to_assets("image_3.png"))
        self.image_3 = canvas.create_image(
            294.0,
            215.0,
            image=self.image_image_3
        )

        self.image_image_4 = PhotoImage(
            file=self.relative_to_assets("image_4.png"))
        self.image_4 = canvas.create_image(
            312.0,
            430.0,
            image=self.image_image_4
        )

        self.image_image_5 = PhotoImage(
            file=self.relative_to_assets("image_5.png"))
        self.image_5 = canvas.create_image(
            413.99990113846434,
            343.0,
            image=self.image_image_5
        )

        # Entries
        self.create_entry(canvas, "Username:", self.var_User_Name, "Enter your user name", 628, 133, "entry_1.png")
        self.create_entry(canvas, "Full Name:", self.var_full_name, "Enter full name", 628, 64, "entry_2.png")
        self.create_entry(canvas, "Email:", self.var_email, "example@gmail.com", 628, 202, "entry_4.png")
        self.create_entry(canvas, "Password:", self.var_password, "Enter a password", 628, 271, "entry_3.png", show="*")
        self.create_entry(canvas, "Confirm Password:", self.var_confirm_password, "Confirm your password", 628, 343, "entry_5.png", show="*")

        # Register Button
        button_image_1 = PhotoImage(file=self.relative_to_assets("button_1.png"))
        Button(
            self,
            image=button_image_1,
            borderwidth=0,
            highlightthickness=0,
            command=self.registration_check,
            bg="#CBD4F5",
            activebackground="#CBD4F5",
            relief="flat"
        ).place(x=611, y=450, width=300, height=45)
        self.register_button_image = button_image_1  # Prevent garbage collection

        # Already have an account?
        canvas.create_text(
            648, 539, anchor="nw",
            text="Yes, I have an account?",
            fill="#000000", font=("Inter", 16 * -1)
        )

        button_image_2 = PhotoImage(file=self.relative_to_assets("button_2.png"))
        Button(
            self,
            image=button_image_2,
            borderwidth=0,
            highlightthickness=0,
            command=lambda: self.controller.show_frame("LoginPage"),
            bg="#CBD4F5",
            activebackground="#CBD4F5",
            relief="flat"
        ).place(x=831, y=539, width=47, height=19)
        self.login_button_image = button_image_2  # Prevent garbage collection

    def create_entry(self, canvas, label, variable, placeholder, x, y, image_path, show=None):
        canvas.create_text(
            x, y, anchor="nw",
            text=label,
            font=("Inter", 16 * -1),
            fill="#000000"
        )
        entry_image = PhotoImage(file=self.relative_to_assets(image_path))
        canvas.create_image(x + 133, y + 41, image=entry_image)
        entry = PlaceholderEntry(
            self,
            textvariable=variable,
            placeholder=placeholder,
            bd=0,
            bg="#FFFFFF",
            fg="#000716",
            highlightthickness=0
        )
        entry.place(x=x, y=y + 25, width=266, height=32)
        self.entry_images = getattr(self, "entry_images", []) + [entry_image]  # Prevent garbage collection


```
